# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_absolute_error

plt.style.use('seaborn-v0_8-darkgrid')
```

## Reading and parsing the data
https://www.kaggle.com/datasets/sumanthvrao/daily-climate-time-series-data/data

```python
# Introducing the date column using 'parse_dates'
df = pd.read_csv('data/DailyDelhiClimateTrain.csv', parse_dates=['date'], index_col='date')
```

```python
df.info()
```

```python
df.head()
```

```python
# Taking only the meantemp to work on it
data = df['meantemp']
```

```python
# Indicating that our data is daily by 'D' to pandas
# This fills the missing days or sets the frequence
data = data.asfreq('D')
```

```python
data.isna().sum()
```

```python
# We have no nan values but if we had we could fill it with a days before values with this function
data = data.ffill()
```

```python
data.head()
```

```python
# Plotting the figure and visualizing data
plt.figure(figsize=(12,5))
plt.plot(data)
plt.title('Real Daily Temp')
plt.ylabel('Temp')
plt.show()
```

## ADFuller Test

```python
result = adfuller(data.dropna())
print(f'ADF Statistic:{result[0]}')
print(f'p-value: {result[1]}')
```

```python
if result[1]>0.05:
    print('Data is not stationary')
else:
    print('Data is stationary')
```

## Training the Model

```python
train_data = data
test_data = pd.read_csv('data/DailyDelhiClimateTest.csv', parse_dates=['date'], index_col='date')
test_data = test_data['meantemp']
test_data = test_data.asfreq('D')
test_data = test_data.ffill()
print(train_data)
print(test_data)
```

```python
# (1,1,1) might not be the fit but it's a good start
# Seasonality: Since there is a yearly seasonality i'll try 12(monthly) or 7 (weekly) 
# Since the data is daily the seasonality is 365 but its a heavy work
# So first check if there is a weekly seasonality
model = SARIMAX(train_data,
               order=(1, 1, 1),
               seasonal_order=(1, 1, 1, 7),
               enforce_stationarity=False,
               enforce_invertibility=False)

model_fit = model.fit(disp=False)
```

```python
print(model_fit.summary())
```

## Forecasting and Compare

```python
predictions = model_fit.forecast(steps=len(test_data))

# Calculating the error
mae = mean_absolute_error(test_data, predictions)
print(f'Mean Absoulte Error(MAE):{mae:.2f}')
```

## Plotting the results

```python
plt.figure(figsize=(14, 7))
# Showing last 100 days
plt.plot(train_data.index[-100:], train_data.values[-100:], label='Training (Past)', alpha=0.5)
plt.plot(test_data.index, test_data.values, label='Real (Test)', color='green', linewidth=2)
plt.plot(test_data.index, predictions, label='Prediction (Model)', color='red', linestyle='--', linewidth=2)

plt.title(f'ARIMA Test on Real Data (Eror Rate: {mae:.2f})')
plt.legend()
plt.show()
```

### Model couldn't learn that Winter comes after Summer so it looked to the last few weeks and thought it will continue ascending we can either turn seasonality into 365 which would take so much time or resample the data so that we have 12 months seasonality and model can easily recognize the winter -> summer

## Resampling Method

```python
# Resampling Training Data
df = pd.read_csv('data/DailyDelhiClimateTrain.csv', parse_dates=['date'], index_col='date')
data = df['meantemp']

# Here we convertin daily information into monthly mean (Resampling)
# 'MS' Month Start (Labeling as first day of the month)
train_monthly = data.resample('MS').mean()
```

```python
# Resampling Test Data
test_df = pd.read_csv('data/DailyDelhiClimateTest.csv', parse_dates=['date'], index_col='date')
test_data = test_df['meantemp']

# Applying the same step to the test data
test_monthly = test_data.resample('MS').mean()
```

```python
print(f'Train data {len(train_monthly)} months')
print(f'Test data {len(test_monthly)} months')
```

```python
# Training Model
# Same as before just changed the 7 into 12 to catch annualy seasonality
model = SARIMAX(train_monthly, 
                order=(1, 1, 1), 
                seasonal_order=(1, 1, 1, 12), 
                enforce_stationarity=False,
                enforce_invertibility=False)
model_fit = model.fit(disp=False)
```

```python
print(model_fit.summary())
```

### Testing and Error Calculation

```python
predictions = model_fit.forecast(steps=len(test_monthly))

# Assigning dates from test data to predictons
predictions = pd.Series(predictions, index=test_monthly.index)
```

```python
# Results and visualizing

# First place the datas side by side to ensure 
comparison = pd.DataFrame({'Actual': test_monthly, 'Predicted': predictions})

# Control: To see if any Nan Values are there
print(f"Nan values before cleaning: {comparison.isna().sum().sum()}")

# Dropping rows contains nan
comparison = comparison.dropna()

# Now calculating on clean data
mae = mean_absolute_error(comparison['Actual'], comparison['Predicted'])
print(f"\n Mean Absolute Error(MAE): {mae:.2f} C")

plt.figure(figsize=(12, 6))

# Training (Last 2 Years)
plt.plot(train_monthly.index[-24:], train_monthly.values[-24:], label='Train (Past)', alpha=0.5)

# Test
plt.plot(comparison.index, comparison['Actual'], label='Real (Test)', color='green', linewidth=2)

# Prediction
plt.plot(comparison.index, comparison['Predicted'], label='Model Forecast', color='red', linestyle='--', linewidth=2, marker='o')

plt.title(f'Temp Forecast: Real vs Model (Hata: {mae:.2f} C)')
plt.legend()
plt.show()
```

##  Conclusion & Key Takeaways

In this project, I built a time series forecasting system to predict climate trends using historical data from Delhi. The journey from raw data to a working **SARIMAX** model involved several critical steps in data engineering and statistical analysis.

### 1. Data Preprocessing & Stationarity
* Initial EDA and the **Augmented Dickey-Fuller (ADF)** test revealed that the raw daily data was too noisy to model annual seasonality ($s=365$) effectively.
* **Solution:** I applied **resampling** to convert the dataset into **monthly averages**. This reduced noise and allowed the model to effectively capture the annual seasonal cycle ($s=12$).

### 2. Model Selection (ARIMA vs. SARIMAX)
* Standard ARIMA models failed to capture the recurring temperature patterns (seasonality).
* I implemented **SARIMAX**, specifically tuning the seasonal components. The statistical results (P-values) confirmed that the **Seasonal Autoregressive (AR.S)** term was the most significant factor (`P < 0.05`), proving that past seasonal trends are the strongest predictor for future temperatures.

### 3. Handling Real-World Data Challenges
* During the evaluation phase, I encountered inconsistencies due to missing values created during the resampling process (`ValueError: Input contains NaN`).
* I implemented a robust validation pipeline that aligns the test data with predictions and drops `NaN` values before calculating performance metrics.

### 4. Results 
* The final model achieved a **Mean Absolute Error (MAE) of ~2.28C** on the test set.
* Visual analysis confirms that the model successfully predicts the warming trend post-January, overcoming the "flat-line" forecast issue encountered in earlier iterations.
